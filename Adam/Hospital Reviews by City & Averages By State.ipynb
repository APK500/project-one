```python
hospital_reviews = "Total Hospital Review - Averages BY STATE.csv"

```

```python
# Use Pandas to read data
hospital_reviews = pd.read_csv(hospital_reviews)
hospital_reviews.head()
```

```python
plt.scatter(hospital_reviews['State'], hospital_reviews['Adams_State Wide Avg Review - (# stars out of 5)'], marker="o", facecolors="red", edgecolors="black")

# Set the upper and lower limits of our y axis
plt.ylim(0,5)

# Set the upper and lower limits of our x axis
plt.xlim(0,10)

# Create a title, x label, and y label for our chart
plt.title("State VS AVG Review (Stars out of 5)")
plt.xlabel("State")
plt.ylabel("AVG Review - Stars out of 5")

# Save an image of the chart and print to screen
plt.savefig("State VS AVG Review of Hospitals.png")
plt.show()
```
