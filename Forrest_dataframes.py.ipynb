```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Create a reference the CSV file desired
csv_path = "City_GDP.csv"

# Read the CSV into a Pandas DataFrame
df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
df.head(20)
```

```python
gdp_df = df.rename(columns = {"Unnamed: 0":"City", "Unnamed: 1":"State", "Percent":"GDP Growth 2018/2017"})
gdp_df.head(20)
```

```python
plt.bar(gdp_df['City'], gdp_df['GDP Growth 2018/2017'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(0,5)

# Give our chart some labels and a tile
plt.title("City GDP Growth Rates")
plt.xlabel("City")
plt.ylabel("GDP Growth Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
# Create a reference the CSV file desired
csv_path = "State_unemploy_rates.csv"

# Read the CSV into a Pandas DataFrame
unemploy_df = pd.read_csv(csv_path, header=1)

# Print the first five rows of data to the screen
unemploy_df.head(20)
```

```python
plt.bar(unemploy_df['State'], unemploy_df['Unemployment Rate'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(1,7)

# Give our chart some labels and a tile
plt.title("State Unemployemnt Rates")
plt.xlabel("City")
plt.ylabel("Unemployment Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
# Create a reference the CSV file desired
csv_path = "Taxes.csv"

# Read the CSV into a Pandas DataFrame
taxes_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
taxes_df.head(20)
```

```python
taxes_df = taxes_df.rename(columns = {"Unnamed: 0":"City", "Unnamed: 1":"State", "Percent":"State & Local Tax Rate"})
taxes_df
```

```python
taxes_df['Taxes_score'] = stats.rankdata(taxes_df['State & Local Tax Rate']*-1)/(1+len(taxes_df['State & Local Tax Rate']))
taxes_df
```

```python
plt.bar(taxes_df['City'], taxes_df['State & Local Tax Rate'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(4,12)

# Give our chart some labels and a tile
plt.title("State & Local Tax Rates")
plt.xlabel("City")
plt.ylabel("Tax Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
#merge gdp_df and unemploy_df
merge_one = pd.merge(gdp_df, unemploy_df, on="State", how="outer")
merge_one
```

```python
merge_two = pd.merge(merge_one, taxes_df, on="City", how="inner")
merge_two
```

```python
merge_two=merge_two.rename(columns = {"State_x":"State"})
merge_two
```

```python
merge_two = merge_two.drop('State_y', 1)
merge_two
```

```python
merge_two = merge_two.drop('Taxes_score', 1)
merge_two
```
