```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Create a reference the CSV file desired
csv_path = "City_GDP.csv"

# Read the CSV into a Pandas DataFrame
df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
df.head(20)
```

```python
gdp_df = df.rename(columns = {"Unnamed: 0":"City", "Unnamed: 1":"State", "Percent":"GDP Growth 2018/2017"})
# gdp_df["City"]=gdp_df["City"].str.lower()
#gdp_df.head(20)
```

```python
gdp_max = max(gdp_df["GDP Growth 2018/2017"])
gdp_min = min(gdp_df["GDP Growth 2018/2017"])
gdp_range = gdp_max-gdp_min
gdp_df["GDP_score"]=1.0
```

```python
for index,row in gdp_df.iterrows():
    score = (row["GDP Growth 2018/2017"]-gdp_min)/gdp_range*10
    score = round(score,1)
    gdp_df.set_value(index,"GDP_score",score)
    
    
```

```python
gdp_df
```

```python
plt.style.use('ggplot')
plt.bar(gdp_df['City'], gdp_df['GDP Growth 2018/2017'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(0,5)

# Give our chart some labels and a tile
plt.title("City GDP Growth Rates")
plt.xlabel("City")
plt.ylabel("GDP Growth Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
# Create a reference the CSV file desired
csv_path = "State_unemploy_rates.csv"

# Read the CSV into a Pandas DataFrame
unemploy_df = pd.read_csv(csv_path, header=1)

# Print the first five rows of data to the screen
unemploy_df.head(20)
```

```python
unemploy_max = max(unemploy_df["Unemployment Rate"])
unemploy_min = min(unemploy_df["Unemployment Rate"])
unemploy_range = unemploy_max-unemploy_min
unemploy_df["Unemp_score"]=1.0
```

```python
for index,row in unemploy_df.iterrows():
    score = abs((row["Unemployment Rate"]-unemploy_max)/unemploy_range*-10)
    score = round(score,1)
    unemploy_df.set_value(index,"Unemp_score",score)
    
    
```

```python
unemploy_df
```

```python
plt.style.use('ggplot')
plt.bar(unemploy_df['State'], unemploy_df['Unemployment Rate'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(1,7)

# Give our chart some labels and a tile
plt.title("State Unemployemnt Rates")
plt.xlabel("City")
plt.ylabel("Unemployment Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
# Create a reference the CSV file desired
csv_path = "Taxes.csv"

# Read the CSV into a Pandas DataFrame
taxes_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
#taxes_df.head(20)
```

```python
taxes_df = taxes_df.rename(columns = {"Unnamed: 0":"City", "Unnamed: 1":"State", "Percent":"State & Local Tax Rate"})
# taxes_df["City"]=taxes_df["City"].str.lower()
taxes_df
```

```python
taxes_max = max(taxes_df["State & Local Tax Rate"])
taxes_min = min(taxes_df["State & Local Tax Rate"])
taxes_range = taxes_max-taxes_min
taxes_df["Tax_score"]=1.0
```

```python
for index,row in taxes_df.iterrows():
    score = abs((row["State & Local Tax Rate"]-taxes_max)/taxes_range*-10)
    score = round(score,1)
    taxes_df.set_value(index,"Tax_score",score)

    
```

```python
plt.style.use('ggplot')
plt.bar(taxes_df['City'], taxes_df['State & Local Tax Rate'], color='r', alpha=0.5, align="center")

# Sets the y limits of the current chart
plt.ylim(4,12)

# Give our chart some labels and a tile
plt.title("State & Local Tax Rates")
plt.xlabel("City")
plt.ylabel("Tax Rate")
plt.xticks(rotation=90)
# Print our chart to the screen
plt.show()
```

```python
final_df = pd.merge(taxes_df,gdp_df,on="City",how="outer")
```

```python
del final_df["State_y"]
```

```python
del final_df["State & Local Tax Rate"]
del final_df["GDP Growth 2018/2017"]
final_df
```

```python
final_df["Unemp_score"] = 1.0
for index,row in unemploy_df.iterrows():
    state = row["State"]
    score = row["Unemp_score"]
    try:
        final_df.set_value(final_df.index[final_df["State_x"]==state][0],"Unemp_score",score)
    except IndexError:
        continue
```

```python
final_df.set_value(8,"Unemp_score",3.6)
```

```python
final_df["Total Score"] = final_df["Tax_score"]+final_df["GDP_score"]+final_df["Unemp_score"]

final_df
```

```python
overall_max = max(final_df["Total Score"])
overall_min = min(final_df["Total Score"])
overall_range = overall_max-overall_min
final_df["Biz Env Score"]=1.0
```

```python
for index,row in final_df.iterrows():
    score = (row["Total Score"]-overall_min)/overall_range*10
    score = round(score,1)
    final_df.set_value(index,"Biz Env Score",score)
```
